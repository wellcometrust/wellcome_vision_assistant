```python
try:
    import Image
except ImportError:
    from PIL import Image
import pytesseract

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
```

```python
def run_ocr(img_path, rotate=None):
    img=Image.open(img_path)
    
    if rotate:
        img = img.rotate(rotate)

    ocr_text = pytesseract.image_to_string(img)
    print(f"OCR text: {ocr_text}")

    plt.imshow(np.asarray(img))

img_path = '../../data/palace_for_the_people.JPG'
run_ocr(img_path)
```

```python
img_path = '../../data/how_can_you.JPG'
run_ocr(img_path)
```

```python
img_path = '../../data/mosaic_article.JPG'
run_ocr(img_path, rotate=-90)
```

```python
img_path = '../../data/non_mosaic_article.JPG'
run_ocr(img_path)
```
