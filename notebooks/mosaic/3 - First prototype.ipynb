```python
try:
    import Image
except ImportError:
    from PIL import Image
import pytesseract

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

from bs4 import BeautifulSoup
import requests

%matplotlib inline
```

```python
img_path = '/Users/sorrosn/Desktop/ocr_images/mosaic_article.JPG'

img=Image.open(img_path)
img = img.rotate(-90)

ocr_text = pytesseract.image_to_string(img)
print(f"OCR text: {ocr_text}")

plt.imshow(np.asarray(img))
```

```python
google_url = 'https://www.google.co.uk/search'
params = {
    'q': ocr_text
}

response = requests.get(google_url, params=params)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

soup.find('h3').a['href'].split('/url?q=')[1].split('&')[0]
```

```python
google_url = 'https://www.google.co.uk/search'
mosaic_url = 'https://mosaicscience.com/'
params = {
    'q': f"site:{mosaic_url} {ocr_text}"
}

response = requests.get(google_url, params=params)

soup = BeautifulSoup(response.text, 'html.parser')
soup.find('h3').a['href'].split('/url?q=')[1].split('&')[0]
```
