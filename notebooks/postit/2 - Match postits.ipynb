```python
import os

from sklearn.metrics.pairwise import paired_distances

IMAGES_PATH = '../../data/postits/'

def find_closest_image(img, after_imgs):
    img_vector = np.asarray(img).reshape(-1)
    
    match_img_name = None
    for img_name, img_to_match in after_imgs.items():
        img_to_match_vector = np.asarray(img_to_match).reshape(-1)
        
        min_img_dim = min(img_vector.shape[0], img_to_match_vector.shape[0])

        img_dist = paired_distances([img_vector[:min_img_dim]], [img_to_match_vector[:min_img_dim]])[0]
        
        if not match_img_name:
            match_img_name = img_name
            min_dist = img_dist
        elif img_dist < min_dist:
            match_img_name = img_name
            min_dist = img_dist
    
    return match_img_name

after_imgs = {}
for img_name in os.listdir(f"{IMAGES_PATH}/after/"):
    img_path = f"{IMAGES_PATH}/after/{img_name}"
    img = Image.open(img_path)

    after_imgs[img_name] = img

for img_name in os.listdir(f"{IMAGES_PATH}/before/"):
    img_path = f"{IMAGES_PATH}/before/{img_name}"
    img = Image.open(img_path)
    
    match_img_name = find_closest_image(img, after_imgs)
    print(img_name, match_img_name)
```
